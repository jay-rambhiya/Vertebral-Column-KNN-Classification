(b) iii.  Select the first 70 rows of Class 0 and the first 140 rows of Class 1 as the training set

```python
df1_train=df[df['class']==0].head(70)
df2_train=df[df['class']==1].head(140)
df1_test=df[df['class']==0].tail(30)
df2_test=df[df['class']==1].tail(70)
```

```python
df_train=pd.concat([df1_train,df2_train])
df_test=pd.concat([df1_test,df2_test])
```

```python
df_train.reset_index(inplace = True)
df_test.reset_index(inplace = True)
```

```python
df_train_X = df_train.loc[:,'pelvic_incidence':'degree_spondylolisthesis']
df_train_Y = df_train.loc[:,'class']
df_test_X = df_test.loc[:,'pelvic_incidence':'degree_spondylolisthesis']
df_test_Y = df_test.loc[:,'class']
```

# c) Classification using KNN on Vertebral Column Data Set

# (c) i. Function to find the k-nearest neighbors with Euclidean metric

```python
def kNearestClassifier(k):
    
    knnClassifier = KNeighborsClassifier(n_neighbors = k, p = 2, metric = 'minkowski')
    knnClassifier.fit(df_train_X, df_train_Y)
    testPredictedY = knnClassifier.predict(df_test_X.values)
    trainPredictedY = knnClassifier.predict(df_train_X.values)
    trainingScoreError = 1 - accuracy_score(df_train_Y, trainPredictedY)
    testingScoreError = 1 - accuracy_score(df_test_Y, testPredictedY)
    
    return (trainingScoreError, testingScoreError)
```

# c) ii. Training and testing errors for different values of k.¶

```python
trainingErrors = []
testingErrors = []
kValues = []

for k in range(208, 0, -3):
    trainingError, testingError = kNearestClassifier(k)
    trainingErrors.append(trainingError)
    testingErrors.append(testingError)
    kValues.append(k)
```

```python
plt.figure(figsize = (30, 15))
plt.plot(kValues, trainingErrors, color = 'red', label = 'Training Accuracy Error')
plt.plot(kValues, testingErrors, color = 'blue', label = 'Testing Accuracy Error')
plt.xticks(np.arange(0, 210, 5))
plt.xlabel('Values of k')
plt.ylabel('Accuracy Error')
plt.legend()
```

# In the above plot, we can see that the accuracy score error is minimum when k is between 1 and 5 approximately. We will run the function again for values 2, 3, 4, and 5 to find the k* which gives minimum accuracy score error.¶

```python
print('Values of k: Accuracy\n')
for k in range(1,6):
    
    print(str(k) + ": " + str(1 - kNearestClassifier(k)[1]))
```

# We get the highest accuracy when k = 4. Therefore, we take k = 4 for further calculations of the confusion matrix.

```python
knn = KNeighborsClassifier(n_neighbors = 4, p = 2, metric = 'minkowski')
knn.fit(df_train_X, df_train_Y)
yPredictedScores = knn.predict(df_test_X)
testingAccuracy = metrics.accuracy_score(df_test_Y, yPredictedScores)
```

```python
confusionMatrix = metrics.confusion_matrix(df_test_Y, yPredictedScores)
TP = confusionMatrix[1, 1]
TN = confusionMatrix[0, 0]
FP = confusionMatrix[0, 1]
FN = confusionMatrix[1, 0]

print('Confusion Matrix: ')
print(confusionMatrix)

print('\n\nTrue Positive Rate: ')
print(TP / (TP + FN))

print('\n\nTrue Negative Rate: ')
print(TN / (TN + FP))

recall = TP / (TP + FN)
print('\n\nPrecision: ')
precision = TP / (TP + FP)
print(precision)

print('\n\nF1-Score: ')
print(2 * (recall * precision) / (recall + precision))
```

```python
nValues = []
for n in range(10, 211, 10):
    nValues.append(n)

def kNearestClassifierforN(n, trainingX, trainingY, testingX, testingY):
    
    knn = KNeighborsClassifier(n_neighbors = n, p = 2, metric = 'minkowski')
    knn.fit(trainingX, trainingY)
    yPredictions = knn.predict(testingX.values)
    accuracy = accuracy_score(testingY, yPredictions)
    error = 1 - accuracy
    return error
```

```python
minKs = []
minErrors = []
df1 = df[df['class'] == 0].head(70)
df2 = df[df['class'] == 1].head(140)

for n in nValues:
    
    print("---------------------------------------------------------------------------------------------------")
    first = df1.head(n // 3)
    second = df2.head(2 * n // 3)
    
    newTrainingData = pd.concat([first, second], axis = 0)
    newTrainingData.reset_index(inplace = True)
    newTrainingData = newTrainingData.loc[:, 'pelvic_incidence' : 'class']
    xTrain = newTrainingData.loc[:, 'pelvic_incidence' : 'degree_spondylolisthesis']
    yTrain = newTrainingData.loc[:, 'class']
    
    minK, minError = None, 2
    errorDict = {}
    for k in range(1, n, 5):
        
        error = kNearestClassifierforN(k, xTrain, yTrain, df_test_X, df_test_Y)
        errorDict[k] = error
        if error < minError:
            minK = k
            minError = error
    
    print(errorDict)
    print("\nWhen N = " + str(n) + " the best test error rate we get is " + str(minError) + " for k = " + str(minK))
    minErrors.append(minError)
```

```python
plt.figure(figsize = (30, 15))
plt.plot(nValues, minErrors, marker = 'o')
plt.xticks(np.arange(0, 220, 10))
plt.xlabel('Training Set Size')
plt.ylabel('Best Test Error Rate')
plt.show()
```

# (d) Using different metrics

# (d) i. A. Using Manhattan Distance as a metric (p=1)

```python
def knnManhattan(n):
    
    knn = KNeighborsClassifier(n_neighbors = n, p = 1, metric = 'minkowski')
    knn.fit(df_train_X, df_train_Y)
    yPredictedScores = knn.predict(df_test_X.values)
    testScore = accuracy_score(df_test_Y, yPredictedScores)
    error = 1 - testScore
    
    return error

minK, minError = None, 1
errors = []
for k in range(1, 197, 5):
    err = knnManhattan(k)
    errors.append([k, err])
    if err <= minError:
        minK = k
        minError = err

print("The errors for all the values of k are:")
print(errors)
print("\nUsing the Manhattan distance as a metric the test error rate is", minError, "when k is", minK)
```

# (d) i. B. Using log10(p) Values

```python
def knnWithPvalues(p):
    
    knn = KNeighborsClassifier(n_neighbors = minK, p = p)
    knn.fit(df_train_X, df_train_Y)
    yPredictedScores = knn.predict(df_test_X.values)
    testScore = accuracy_score(df_test_Y, yPredictedScores)
    error = 1 - testScore
    
    return error

minP, minError = None, 1
errors = []
for logp in np.arange(0.1, 1.1, 0.1):
    p = 10 ** logp
    err = knnWithPvalues(p)
    errors.append([logp, err])
    if err <= minError:
        minP = logp
        minError = err

print("The errors for all the values of logp are:")
print(errors)
print("\nUsing the Manhattan distance as a metric the test error rate is", minError, "when logp is", minP)
```

# (d) ii. Using Mahalanobis Distance as a metric

```python
def knnMahalanobis(n):
    
    knn = KNeighborsClassifier(n_neighbors = n, metric = 'mahalanobis', metric_params = {'VI': np.linalg.pinv(np.cov(df_train_X, rowvar = False))}, algorithm = 'brute')
    knn.fit(df_train_X, df_train_Y)
    yPredictedScores = knn.predict(df_test_X.values)
    testScore = accuracy_score(df_test_Y, yPredictedScores)
    error = 1 - testScore
    
    return error

minK, minError = None, 1
errors = []
for k in range(1, 197, 5):
    err = knnMahalanobis(k)
    errors.append([k, err])
    if err < minError:
        minK = k
        minError = err

    
print("The errors for all the values of k are:")
print(errors)
print("\nUsing the Mahanalobis distance as a metric the test error rate is", minError, "when k is", minK)
```
